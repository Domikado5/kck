# Zadania - Sympy

```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
x, y = sym.symbols("x y")
```

```python
f = -x**3 + 3*x**2 + 10*x - 24
solve_points = np.array(sym.solve(f))
data = {'x': np.arange(-5, 5.1, 0.1),
'y': [f.subs(x, step) for step in np.arange(-5, 5.1, 0.1)]}

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100
fig, ax = plt.subplots(1, 1)
ax.set_title("Wielomian -x^3 + 3x^2 + 10x - 24")
ax.scatter(solve_points, [f.subs(x, step) for step in solve_points])
ax.plot(data['x'], data['y'])
ax.axis(xmin=-5, xmax=5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xticks([i for i in range(-5, 6, 1)])
ax.legend(['-x^3 + 3x^2 + 10x - 24'], numpoints=2)
ax.grid(color='gray', linestyle='-', dashes=(2,6), linewidth=0.45)
ax.tick_params(direction='in')
plt.show()
plt.close()
```

```python
from sympy.interactive import printing
printing.init_printing(use_latex=True)
import sympy as sym

x, y = sym.symbols("x y")
sol = sym.solve([x**2+3*y-10, 4*x-y**2+2])
sol
```

```python
print(f"Liczba rozwiza: {len(sol)}")
```

```python
for s in sol:
    for k, v in s.items():
        print(f"{k}: {v.evalf()}", end=' ')
    print('')
```

```python
x = [
    np.arange(-np.pi, np.pi+np.pi, np.pi),
    np.arange(-np.pi, np.pi+2*np.pi/10, 2*np.pi/10),
    np.arange(-np.pi, np.pi+2*np.pi/100, 2*np.pi/100)
]
y = [
    np.sin(i) for i in x
]
legend = ['', '2/10', '2/100', 'wszystkie']
color = ['firebrick', 'cornflowerblue', 'goldenrod']
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2)

step = 0
for i in range(2):
    for j in range(2):
        if step < 3:
            axs[i, j].plot(x[step], y[step], color=color[step])
            axs[1, 1].plot(x[step], y[step], color=color[step])
        axs[i, j].axis(xmin=-np.pi, xmax=np.pi, ymin=-1, ymax=1)
        axs[i, j].set_xlabel("x")
        axs[i, j].set_ylabel("y")
        axs[i, j].set_xticks(np.arange(-np.pi, np.pi*1.5, np.pi/2))
        axs[i, j].set_xticklabels(['-', '-/2', '0','/2', ''])
        axs[i, j].grid(color='gray', linestyle='-', dashes=(2,6), linewidth=0.45)
        axs[i, j].tick_params(direction='in')
        axs[i, j].set_title(f'Wykres funkcji sin (krok {legend[step]})' if step < 3 else f'Wykresy funkcji sin', fontsize=16)
        axs[i, j].legend(['', '2/10', '2/100']) if step == 3 else None
        step += 1

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100
fig.tight_layout(pad=3.0)

plt.show()
plt.close()
```
